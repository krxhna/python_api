```python
import requests 
from bs4 import BeautifulSoup
```

```python
user_agent = {'User-agent': 'Mozilla/5.0'}

doc = requests.get('https://www.sec.gov/ix?doc=/Archives/edgar/data/789019/000156459021039151/msft-10k_20210630.htm' ,headers=user_agent).text
```

```python
soup = BeautifulSoup(doc, 'html.parser')

print(soup.prettify())
```
